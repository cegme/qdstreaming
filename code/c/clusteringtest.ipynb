```python
filename = "sortedtest6.csv"

# Read params
params = ""
with open(filename, 'r') as f:
  params = f.readline()

print(params)
    
# Read data
test = pd.read_csv(filename, header=1)
test.head()
```

```python
# Get the data
xbaseline = filter(lambda x: x[1] == "BASELINE", test.itertuples())
xsorted = filter(lambda x: x[1] == "SORTED", test.itertuples())

ind = np.arange(len(test)/2)
width = 0.5

# Ticks
plt.xticks(ind+width/2, map(lambda x: (x[3],x[4]), xsorted), rotation=90) 

# Lables
plt.title("Study of clustering and early stopping methods")
plt.ylabel("Time (secs * 1000000)")
plt.xlabel("(Left Cluster, Right Cluster)")

plt.annotate(params, xy=(0,200), xycoords='data', textcoords='offset points')
p1 = plt.bar(ind, map(lambda x: x[5], xbaseline), width, color='#f1a340',\
             yerr=map(lambda x: sqrt(float(x[6])), xbaseline), log=True)
p2 = plt.bar(ind+width, map(lambda x: x[5], xsorted), width, color='#998ec3',\
             yerr=map(lambda x: sqrt(float(x[6])), xsorted), log=True)
plt.legend((p1[0], p2[0]), ('baseline', 'sorted'), loc=4)
plt.grid(axis = 'y', color ='white', linestyle='-')

```

```python
def buildchart(filename):
    
    # Read params
    params = ""
    with open(filename, 'r') as f:
      params = f.readline()
    
    print(params)
        
    # Read data
    test = pd.read_csv(filename, header=1)
    test.head()
    
    # Get the data
    xbaseline = filter(lambda x: x[1] == "BASELINE", test.itertuples())
    xsorted = filter(lambda x: x[1] == "SORTED", test.itertuples())
    xbase_triangle = filter(lambda x: x[1] == "BASELINE_TRIANGLE", test.itertuples())
    
    
    ind = np.arange(len(test)/3)
    width = 0.25
    
    # Ticks
    plt.xticks(ind+width/2, map(lambda x: (x[3],x[4]), xsorted), rotation=90) 
    
    # Lables
    plt.title("Study of clustering and early stopping methods")
    plt.ylabel("Time (secs * 1000000)")
    plt.xlabel("(Left Cluster, Right Cluster)")
    
    plt.annotate(params, xy=(0,200), xycoords='data', textcoords='offset points')
    
    p1 = plt.bar(ind, map(lambda x: x[5], xbaseline), width, color='#f1a340',\
                 yerr=map(lambda x: sqrt(float(x[6])), xbaseline), log=True)
    p2 = plt.bar(ind+width, map(lambda x: x[5], xsorted), width, color='#998ec3',\
                 yerr=map(lambda x: sqrt(float(x[6])), xsorted), log=True)
    p3 = plt.bar(ind+width+width, map(lambda x: x[5], xbase_triangle), width, color='red',\
                 yerr=map(lambda x: sqrt(float(x[6])), xbase_triangle), log=True)
    plt.legend((p1[0], p2[0], p3[0]), ('baseline', 'sorted', 'baseline_triangle'), loc=4)
    plt.grid(axis = 'y', color ='white', linestyle='-')
```

```python
buildchart('sortedtest6.csv')
```
