```python
def buildchart(filename):
    
    # Read params
    params = ""
    with open(filename, 'r') as f:
      params = f.readline()
    
    print(params)
        
    # Read data
    test = pd.read_csv(filename, header=1)
    test.head()
    
    # Get the data
    #xbaseline = filter(lambda x: x[1] == "BASELINE", test.itertuples())
    xsorted = filter(lambda x: x[1] == "SORTED", test.itertuples())
    xbase_triangle = filter(lambda x: x[1] == "BASELINE_TRIANGLE", test.itertuples())
    xblocking = filter(lambda x: x[1] == "BLOCKING", test.itertuples())
    
    
    ind = np.arange(len(test)/5)
    width = 0.20
    
    # Ticks
    plt.xticks(ind+width/3, map(lambda x: (x[3],x[4]), xsorted), rotation=90) 
    
    # Lables
    plt.title("Study of clustering and early stopping methods")
    plt.ylabel("Time (secs * 1000000)")
    plt.xlabel("(Left Cluster, Right Cluster)")
    
    plt.annotate(params, xy=(0,225), xycoords='data', textcoords='offset points')
    
    #p1 = plt.bar(ind, map(lambda x: x[5], xbaseline), width, color='#f1a340',\
    #             yerr=map(lambda x: sqrt(float(x[6])), xbaseline), log=True)
    p1 = plt.bar(ind+1*width, map(lambda x: x[5], xsorted), width, color='#CC6666',\
                 yerr=map(lambda x: sqrt(float(x[6])), xsorted), log=True)
    p2 = plt.bar(ind+2*width, map(lambda x: x[5], xbase_triangle), width, color='#9999CC',\
                 yerr=map(lambda x: sqrt(float(x[6])), xbase_triangle), log=True)
    p3 = plt.bar(ind+3*width, map(lambda x: x[5], xblocking), width, color='#66CC99',\
                 yerr=map(lambda x: sqrt(float(x[6])), xblocking), log=True)
    
    plt.legend((p1[0], p2[0], p3[0]), ('sorted', 'baseline', 'blocking'), loc=2)
    plt.grid(axis = 'y', color ='white', linestyle='-')

buildchart('sortedtest7.csv')
```
